```python
def get_videos(parent_folder):
    """Lấy tất cả video từ 'videos_info.json' trong parent_folder, sau đó lọc video có view > view_threshold."""

    all_videos = []

    # Kiểm tra thư mục tồn tại
    if not os.path.exists(parent_folder):
        print(f"Error: Folder '{parent_folder}' does not exist.")
        return []
    subfolder_count = 0
    # Duyệt qua tất cả sub-folder trong parent_folder
    for root, dirs, files in os.walk(parent_folder):
        subfolder_count += 1
        if "videos_info.json" in files:  # Chỉ xử lý 'videos_info.json'
            file_path = os.path.join(root, "videos_info.json")

            try:
                with open(file_path, "r", encoding="utf-8") as file:
                    data = json.load(file)

                    # Đảm bảo dữ liệu là list chứa dictionary
                    if isinstance(data, list) and all(isinstance(item, dict) for item in data):
                        all_videos.extend(data)  # Gom tất cả video lại

            except (json.JSONDecodeError, ValueError):
                print(f"Skipping {file_path}: Invalid JSON format.")
            except Exception as e:
                print(f"Error reading {file_path}: {e}")
    print(f"Found {len(all_videos)} videos in {subfolder_count} sub-folders.")
    return all_videos
```

```python
# Example usage
parent_folder = "data/raw/user_amthuc"  # Thay bằng đường dẫn thư mục của bạn
all_videos = get_videos(parent_folder)
```

```python
# Lọc video có 'playCount' > view_threshold
view_threshold = 1000000
high_view_videos = []
for video in all_videos:
    stats = video.get("stats", video.get("statsV2", {})
                      )  # Lấy 'stats' hoặc 'statsV2'
    play_count = stats.get("playCount", 0)  # Nếu không có, mặc định là 0

    if play_count > view_threshold:
        high_view_videos.append(video)
        print(video.get("stats"))

print(f"\n📊 Total videos found: {len(all_videos)}")
print(f"🔥 Videos with playCount > {view_threshold}: {len(high_view_videos)}")
```
