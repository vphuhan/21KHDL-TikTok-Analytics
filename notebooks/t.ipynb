```python
import pandas as pd

df = pd.read_parquet('preprocessed_videos.parquet')
```

```python
df.columns
```

```python
user_stats = df.sort_values('authorStats.followerCount', ascending=False).groupby(
    'author.uniqueId').size().reset_index(name='videos_count')
fl_df = df.groupby('author.uniqueId')[
    'authorStats.followerCount'].max().reset_index(name='followers')
user_stats.merge(fl_df, on='author.uniqueId').sort_values(
    'followers', ascending=False).head(50)
```
