```python
df = pd.read_csv('data/videos_with_weekly_rank.csv', dtype={'id': object})
```

```python
top_50_df = df[['year', 'week', 'id', 'weekly_score_rank', 'desc']]
top_50_df = top_50_df.sort_values(['year', 'week', 'weekly_score_rank'])
```

```python
top_50_df = top_50_df.groupby(['year', 'week']).head(50).drop_duplicates()
```

```python
top_50_df = top_50_df.rename(columns={'id': 'video.id'})
```

```python
# ƒê∆∞·ªùng d·∫´n folder ch·ª©a JSON
json_folder = 'data/transcripts-15K'

# Danh s√°ch l∆∞u k·∫øt qu·∫£
json_data_list = []

# L·∫∑p qua t·ª´ng ID trong DataFrame
for video_id in top_50_df['video.id']:
    file_path = os.path.join(json_folder, f"{video_id}.json")
    if not os.path.exists(file_path):
        # print(f"‚ö†Ô∏è File kh√¥ng t·ªìn t·∫°i: {file_path}")
        continue

    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            content = f.read()        # Th√™m video.id v√†o n·ªôi dung JSON ƒë·ªÉ merge ƒë∆∞·ª£c
        # lo·∫°i b·ªè k√Ω t·ª± ƒëi·ªÅu khi·ªÉn
        content = re.sub(r'[\x00-\x1F\x7F]', '', content)
        content = json.loads(content)
        content['video.id'] = video_id
        json_data_list.append(content)
    except Exception as e:
        # print(f"‚ùå L·ªói khi ƒë·ªçc file {file_path}: {e}")
        continue

# Chuy·ªÉn th√†nh DataFrame t·ª´ danh s√°ch JSON
json_df = pd.DataFrame(json_data_list)
# json_df = json_df.merge(
#     transcript_df[['video.id', 'transcript']], on='video.id')
json_df.info()
```

```python
json_df = json_df.merge(top_50_df, on='video.id')
```
