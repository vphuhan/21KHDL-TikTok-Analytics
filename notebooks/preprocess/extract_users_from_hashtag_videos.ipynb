```python
# Lọc lại các cột quan trọng liên quan đến user và các chỉ số tương tác của video
df_selected = df[[
    # "author.id",  # ID của user
    "author.uniqueId",  # Tên hiển thị của user
    'authorStats.followerCount',
    "statsV2.playCount",  # Lượt xem video
    "statsV2.diggCount",  # Lượt thích video
    "statsV2.commentCount",  # Lượt bình luận video
    "statsV2.shareCount",  # Lượt chia sẻ video
    'statsV2.repostCount',
    'statsV2.collectCount'
]]

# Kiểm tra dữ liệu bị thiếu
df_selected = df_selected.dropna()

# Tổng hợp dữ liệu theo user
df_grouped0 = df_selected.groupby(["author.uniqueId"]).agg(
    total_views=("statsV2.playCount", "sum"),
    avg_views=("statsV2.playCount", "mean"),
    total_likes=("statsV2.diggCount", "sum"),
    avg_likes=("statsV2.diggCount", "mean"),
    total_comments=("statsV2.commentCount", "sum"),
    total_shares=("statsV2.shareCount", "sum"),
    total_collects=("statsV2.collectCount", "sum"),
    total_reposts=("statsV2.repostCount", "sum"),
    total_videos=("statsV2.playCount", "count"),  # Số lượng video của user
    followers=('authorStats.followerCount', "max")
).reset_index()

# Tính Engagement Rate (ER) trung bình
df_grouped0["engagement_rate"] = (df_grouped0["total_likes"] + df_grouped0["total_comments"] +
                                  df_grouped0["total_shares"]+df_grouped0["total_collects"] +
                                  df_grouped0["total_reposts"]) / df_grouped0["total_views"]

# Loại bỏ những user có tổng lượt xem bằng 0 (tránh lỗi chia cho 0)
df_grouped0 = df_grouped0[df_grouped0["total_views"] > 0]
df_grouped0.shape
```

```python
# Xác định ngưỡng tối thiểu để loại bỏ user không hoạt động mạnh trong lĩnh vực ẩm thực
# min_videos_threshold = df_grouped["total_videos"].quantile(
#     0.80)  # Loại bỏ user có số video thấp hơn top 25%
min_total_views_threshold = df_grouped["total_views"].quantile(
    0.25)  # Loại bỏ user có tổng lượt xem thấp hơn top 25%
min_views_threshold = df_grouped["avg_views"].quantile(
    0.25)  # Loại bỏ user có tổng lượt xem thấp hơn top 25%
min_er_threshold = df_grouped["engagement_rate"].quantile(
    0.5)  # Loại bỏ user có engagement thấp hơn top 25%

# Lọc lại danh sách user, chỉ giữ lại những người có ít nhất một tiêu chí vượt ngưỡng tối thiểu
df_filtered_users = df_grouped[
    # (df_grouped["total_videos"] >= 4) &
    (
        (df_grouped["avg_views"] >= min_views_threshold)
        &
        (df_grouped["total_views"] >= min_total_views_threshold)
        &
        (df_grouped["engagement_rate"] >= min_er_threshold)
    )
]
print(df_filtered_users.shape)
display(df_filtered_users.drop(columns=['total_comments', 'total_shares', 'total_collects', 'total_reposts'])
        .sort_values(by="total_videos", ascending=False).head(40))
# filtered_user = list(df_filtered_users['author.uniqueId'].unique())
```
